```python
df3 = df
df3.size
```

```python
#number of locations that increased in value or decreased in value each month



df4 = df3.dropna(subset=['2019-10-31','2019-11-30','2019-12-31','2020-01-31','2020-02-29','2020-03-31'])
df4.size

```

```python
#apply function on row
def test_func(row):
    if row['2019-11-30'] > row['2019-10-31']:
        row['delta 2019-11'] = 1
    else:
        row['delta 2019-11'] = 0
    if row['2019-12-31'] > row['2019-11-30']:
        row['delta 2019-12'] = 1
    else:
        row['delta 2019-12'] = 0
    if row['2020-01-31'] > row['2019-12-31']:
        row['delta 2020-01'] = 1
    else:
        row['delta 2020-01'] = 0
    if row['2020-02-29'] > row['2020-01-31']:
        row['delta 2020-02'] = 1
    else:
        row['delta 2020-02'] = 0
    if row['2020-03-31'] > row['2020-02-29']:
        row['delta 2020-03'] = 1
    else:
        row['delta 2020-03'] = 0
    return row

df5 = df4.apply(test_func, axis='columns')
df5
```

```python
#delta = df5[['delta 2019-11','delta 2019-12','delta 2019-12','delta 2019-12','delta 2019-12']]
delta = df5.iloc[:,[5,-1,-2,-3,-4,-5]]
delta
```

```python
delta.sum() #number of locations that increased in value m.o.m dropped in march
```

```python
#breakdown of % by state
states = delta.groupby('State').sum()
states = states.reset_index()
states
```

```python
total = delta.groupby('State').count()
total
```

```python
states = states.merge(total['delta 2019-11'], left_on='State', right_on='State')
states = states.rename(columns={"delta 2019-11_x": "delta 2019-11", "delta 2019-11_y": "total"})
states
```

```python
states['percent'] = ((states['delta 2019-11'] - states['delta 2019-12'])/states['total'])
pd.options.display.max_rows = 100
states.sort_values('percent', ascending=False)

# CA, NY saw decrease in values, TX saw increase than then normal amount of increases/decreases
```
