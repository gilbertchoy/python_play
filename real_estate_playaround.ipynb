```python
import pandas as pd
pd.options.display.max_columns = 9999
pd.options.display.max_rows = 10
df = pd.read_csv('Zip_Zhvi_4bedroom.csv', encoding='latin')
```

```python
df.head()
```

```python
def growth(row):
    if ( (float(row['2020-01']) > (row['2019-01'] * 1.05) ) & (row['2019-01'] > (row['2018-01'] * 1.05)) ):
        row['growing'] = 1
    else:
        row['growing'] = 0
    return row

df_growing = df.apply(lambda row: growth(row), axis='columns')

df_growing = df_growing[df_growing['growing'] == 1]
```

```python
df_growing
```

```python
growing_states = df_growing.groupby('State')['growing'].agg([sum, len])
growing_states = growing_states.reset_index()
growing_states = growing_states[['State','len']]
growing_states = growing_states.rename(columns = {'len':'growingcount'})
growing_states
```

```python
state_all = df.groupby('State').size()
state_all
dfs = state_all.reset_index()
dfs = dfs.rename(columns = {0:'fullcount'})
```

```python
dfstate = dfs.merge(growing_states, left_on = 'State', right_on = 'State')
dfstate
```

```python
dfstate['growingratio'] = dfstate['growingcount']/dfstate['fullcount']
dfstate.sort_values('growingratio', ascending=True)
dfstate
```
