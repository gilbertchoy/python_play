```python
import pandas as pd
pd.options.display.max_columns = 9999
pd.options.display.max_rows = 999
df = pd.read_csv('Zip_Zhvi_4bedroom.csv', encoding='latin')
```

```python
df.head()
```

```python
def growth(row):
    if ( (float(row['2020-01']) > (row['2019-01'] * 1.05) ) & (row['2019-01'] > (row['2018-01'] * 1.05)) ):
        row['growing'] = 1
    else:
        row['growing'] = 0
    return row

df1 = df.apply(lambda row: growth(row), axis='columns')

df1 = df1[df1['growing'] == 1]
```

```python
df1.count()
```

```python
pd.options.display.max_rows = 999
state = df1.groupby('State').size()
state
```

```python
df2 = df1[df1['State'] == 'CA']
df2 = df2.sort_values('CountyName')
df2
```
