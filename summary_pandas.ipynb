```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2'],
                      'C': [0, 1, 2]})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
```

```python
#merge will merge dataframes on whatever column you want
left.merge(right, left_on='key', right_on='key1')
left.merge(right, left_on='C', right_index=True)
#join always uses index values, if want to join on other columns must change columns to index beforehand
left.join(right, lsuffix='lkey', rsuffix='rkey') #the suffixes only matter when both dataframes have same column names
```
